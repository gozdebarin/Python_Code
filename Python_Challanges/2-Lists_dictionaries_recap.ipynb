Lists

```python
# Define a list
list = ["item_1", "item_2"]
```

```python
# Show the list
list
```

```python
# Append an element to the list
list.append(["item_3", "item_7"])
```

```python
# Show the list
list
```

```python
# Insert a new value into the list
list.insert(2, [4, "item_5"])
```

```python
# Show the list
list
```

```python
# Get the index of "item_2"
list.index("item_2")
```

```python
# Get the element that is 3rd in the big list and 2nd in the nested list
list[2][1]
```

```python
# Set the element with index -3 to be equal to "item_6"
list[-3] = "item_6"
```

```python
# Show the list
list
```

Dictionaries

```python
# Create a dictionary
dict = {"key_3":3, "key_2":2}
```

```python
# Show the dictionary
dict
```

```python
# Add key-value pairs to the dictionary
dict.update({"key_1":1, "key_2":2})
```

```python
# Show the dictionary
dict
```

```python
# Add more key-value pairs to the dictionary
dict.update({"key_4":"alligator", "key_5":[200, "dinosaur", True]})
```

```python
# Show the dictionary
dict
```

```python
# Get the value for key "key_4"
dict["key_4"]
```

```python
# Change the value(s) of key "key_2"
dict.update({"key_2":[5, 2]})
```

```python
# Show the dictionary
dict
```

```python
# Get the value for key "key_5"
# This is a list. Get the element with index 2 from that list
dict["key_5"][2]
```

```python
# Same as above
dict.get("key_5")[2]
```
