## 3. What's the average quantity of titles sold per sale by store 7131?

````python
(
    sales['qty']
 [sales["stor_id"]==7131].mean()
)

````

### Hint:

In SQL the syntax is:

```sql
SELECT AVG(qty)
FROM sales

```

## 4. When was the employee with the highest level hired?

````python
employee[["hire_date", "job_lvl"]].sort_values(by="job_lvl", ascending = False).head(1)
````

### Hint:

In SQL the syntax is:

```sql
SELECT hire_date
FROM employee

LIMIT 1;
```

## 5. What's the average price of psychology books?

````python
titles["price"] [titles["type"]=="psychology"].mean()
````

### Hint:

```sql

FROM titles
```

## 7. What's the title and the price of the most expensive book?
